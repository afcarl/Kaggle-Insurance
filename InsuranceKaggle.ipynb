```python
param_grid = {
                'n_estimators': [ 175, 200],
                'max_depth': [9, 19],
                'max_leaf_nodes' : [5 ,15],
                'min_samples_leaf' : [100,500]
            }
rf_model_tuning = RandomForestClassifier(n_jobs=-1, random_state =50)
grid_clf = GridSearchCV(rf_model_tuning, param_grid, cv=4)
grid_clf.fit(train.drop(['id', 'target'], axis=1), train.target)
```

```python
grid_clf.best_params_
```

```python
rf_model_tuning = RandomForestClassifier(n_jobs=-1, oob_score=True, max_depth = grid_clf.best_params_['max_depth'], max_leaf_nodes = grid_clf.best_params_['max_leaf_nodes'], min_samples_leaf = grid_clf.best_params_['min_samples_leaf'], n_estimators = grid_clf.best_params_['n_estimators'])
rf_model_tuning.fit(train.drop(['id', 'target'], axis=1), train.target)
rf_model_tuning.score(train.drop(['id', 'target'], axis=1), train.target)
```
